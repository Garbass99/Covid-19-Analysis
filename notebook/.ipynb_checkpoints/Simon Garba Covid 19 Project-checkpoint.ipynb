```python
def catFeatSelct(data):
    catVars = list(data.select_dtypes('object').columns) # extracts the column names for categorical features and converts to a list
    catDf = data.select_dtypes('object') # extracts sub data that are ONLY categorical as a new dataframe
    return catVars, catDf



def msLblDetect(colNames, catData):
    lblSummary = [] # empty list to contain columns namesand their uniques labels

    for col in colNames:
        colName = col
        uniqLabs = catData[col].unique() # the unique labels in the given columns
        uniqLabs = list(uniqLabs)
        lblSummary.append([colName,uniqLabs ])
    print(lblSummary)
    return lblSummary
```

#Treating Errors

```python
#Treating Missing Values
df['NEW CASES'] = df['NEW CASES'].fillna(df['NEW CASES'].median()) 
df['NEW_DEATHS'] = df['NEW_DEATHS'].fillna(df['NEW_DEATHS'].median())
df['VACCINATED'] = df['VACCINATED'].fillna(df['VACCINATED'].median())
```

```python
#Converting Date to dateTime data type 
df.loc[:, 'DATE'] = pd.to_datetime(df["DATE"])
```

```python
#Dropping Outliers
df = df.drop_duplicates()
```

```python
#sergregating categorica and continues variable
#def dtypesseparator(dfn_d):
Cont_vers = df.select_dtypes('float')
cat_ver = df.select_dtypes('object')
    
```

```python
#Treating ouliers
def outliers_limit(col): #creating a function to determing the upper limit and the lower limit of our IQR
    q3,q1 = np.percentile (col, [75,25])
    iqr = q3-q1
    upper_limit = q3 + 1.5*iqr
    Lower_Limit = q1 - 1.5*iqr
    return upper_limit,Lower_Limit
```

```python
for col in Cont_vers:
    print("----------------------------------------------")
    print("Column:", col)
    UL,LL = outliers_limit (df[col])
    print("Upper Limit = ",UL)
    print("Lower Limit = ",LL)
    total_outliers = len(df.loc[df[col]<LL,col]) + len(df.loc[df[col]>UL,col])
    percent = (total_outliers/len(df.index))*100
    print("percentage of outliers=",percent )
    print("----------------------------------------------/n")
```

```python
"""Observations: New Cases and New Deaths has an outliers that needs to be treated befor continuing with our Analysis
reason is there are other data points that fall far abway from the central tendancy and the rule of distribution stated that every data has an equal
chance of occuring in a sample"""
```

```python
#Treating Outlies in the columns affected
for col in list(Cont_vers):
    UL, LL = outliers_limit(df[col])
    df.loc[df[col] < LL, col]=np.nan
    df.loc[df[col] > UL, col]=np.nan
df.dropna(subset = ['NEW CASES','NEW_DEATHS',], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
#create a function for categorical analysis

def cat_univariate_analysis(Catg_df):
    for col_name in Catg_df:
        print("*"*10, col_name, "*"*10)
        print(Catg_df[col_name].agg(['count', 'nunique', 'unique']))
        print()
        print('Value Counts: \n', Catg_df[col_name].value_counts())
        print()

cat_univariate_analysis(Categ_val)
```

```python
#function for numerical analysis

def numerical_univariate_analysis(numerical_df):
    for col_name in numerical_df:
        print("*"*10, col_name, "*"*10)
        print(numerical_df[col_name].agg(['min', 'max', 'mean', 'median', 'std', 'kurt', 'skew']))
        print()

numerical_univariate_analysis(Contn_val)
```

```python
#creating a function to plot the distribution of the variables
def univariate_analysis(Cont_vers):
    num_cols = len(Cont_vers.columns)
    fig, axes = plt.subplots(nrows=(num_cols // 3) + 1, ncols=3,
                             figsize=(18, 4 * ((num_cols // 3) + 1)))
    fig.tight_layout(pad=5.0)

    axes = axes.flatten()  # Flatten axes array for easy iteration

    for i, col_name in enumerate(Cont_vers.columns):
        sn.histplot(data=Cont_vers, x=col_name, kde=True, ax=axes[i])
        # Use KDE for a smooth distribution curve
        axes[i].set_title(f'Distribution of {col_name}')
        axes[i].set_xlabel(col_name)
        axes[i].set_ylabel('Frequency')

    # Hide any extra subplots if the number of numeric columns is not a multiple of 3
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
        path = r"C:\Users\DELL\Desktop\3MTT Data Science\Covid19_Analysis\visualization"
        filename = "Box_plot_distribution"
        savepath = (path +'/'+filename+'.png')
        plt.savefig(savepath)

    plt.show()

univariate_analysis(Cont_vers)
```

```python
#Trend of total sales by product to identify trend of days, products sold and total price they sold at
import matplotlib.dates as mdates

def covid_trends(df, country=None):
    """ COVID-19 trends for new cases, deaths, and vaccinations"""
    
    # Filter by country if specified
    if country:
        df = df[df['COUNTRY'] == country].copy()
        title = f'COVID-19 Trends in {country}'
    else:
        title = 'Total COVID-19 Trends accross Argentina Austriala, Brazil, Canada and China with New Cases Morthan Vaccinated'
    
    # Convert date and sort
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.sort_values('DATE', inplace=True)
    
    # Create figure
    plt.figure(figsize=(12, 6))
    
    # Plot all three metrics
    plt.plot(df['DATE'], df['NEW CASES'], 'b-', label='New Cases')
    plt.plot(df['DATE'], df['NEW_DEATHS'], 'r-', label='New Deaths')
    plt.plot(df['DATE'], df['VACCINATED'], 'g-', label='Vaccinated')
    
    # Formatting
    plt.title(title)
    plt.ylabel('Count')
    plt.xlabel('Date')
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    # Format x-axis
    ax = plt.gca()
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
covid_trends(Clean_df, country=None)
```

```python
''' Observation: The trend above shows the trend of people vaccinated, New cases identified and the new Death by Covid 19 for the reported period
In the trend we have New Cases identified morethen Vaccination accross all dates with minimal number of death'''
```
